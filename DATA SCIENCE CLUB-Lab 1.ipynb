```python
#Importing the required module from python programming environment
import pandas as pd
```

```python
#Reading the csv file from local storage and storing it in a variable 'df'
#Note that it will vary depending on where your dataset is stored
df=pd.read_csv("F:\\Datasets for DS club\\AB_NYC_2019.csv")
```

```python
#Function to show the first four rows of the dataset to try and understand it's contents
df.head()
```

```python
#Droping Nan values from the dataset(Data cleaning)
df.dropna()
```

```python
'''
Q1.List all properties by their id and property name that are within Chelsea neighbourhood.
-First step  was to filter only properties within Chelsea neighbourhood.
-This was done by creating a condition that checked for the string 'Chelsea' in every row.
-If condition is true, then it displays that row.If false then it skips that row
-The result is later stored in a variable called Chelsea.
'''
Chelsea = df[df["neighbourhood"]=='Chelsea']
```

```python
#Last step was to list only id and property name columns.
Chelsea[['id',"name"]]
```

```python
'''
Q2.List all properties by name, room type and availability, that received less than 50 reviews.
-First step  was to filter only properties with less than 50 reviews
-This was done by creating a condition that checked the 'number_of_reviews' column for any number less than 50
-If condition is true, then it displays that row.If false then it skips that row
-The result is stored in a variable called Reviews_below_50
'''
Reviews_below_50 = df[df['number_of_reviews'] < 50]
```

```python
#Last step was to list only name, room type and availability columns.
Reviews_below_50[['name', 'room_type', 'availability_365']]
```

```python
'''
Q3.List all properties by name, room type and availability, that received less than 250 reviews.
-First step  was to filter only properties with less than 250 reviews
-This was done by creating a condition that checked the 'number_of_reviews' column for any number less than 50
-If condition is true, then it displays that row.If false then it skips that row
-The result is stored in a variable called Reviews_below_250
'''
Reviews_below_250 = df[df['number_of_reviews'] < 250]
```

```python
#Last step was to list only name, room type and availability columns.
Reviews_below_250[['name', 'room_type', 'availability_365']]
```

```python
'''
Q4.List the property by name, room type and availability, that received the highest reviews.
-First step was to use the function max() to try and find the maximum value in the column 'number_of_reviews'
'''
df['number_of_reviews'].max()
```

```python
#Since now we know the highest review is 629, we can use this information to display only that row
#We then save this information in a variable called Highest_review
Highest_review = df[df['number_of_reviews'] == 629]
```

```python
#Last step is to list only name, room type and availability columns
Highest_review[['name', 'room_type', 'availability_365']]
```

```python
'''
Q5.How many properties do each host have?
-We solved this by using the groupby function to group the host's names
-We then used the count function to count the number of times a host has been repeated
-This is also because we noticed every row represents a property 
-For example if Jennifer has been repeated 154 times then that means she has 154 properties
'''
hosts_properties = df['name'].groupby(df['host_name']).count()
```

```python
#We then used the head() function to display the first 5 hosts
print("A list of hosts and the number of properties they have")
hosts_properties.head()
```

```python
#I decided to create a function to check the number of properties a specific host has
#This is because I noticed the list might be too long for someone to find a specific host
#The function takes only one argurement which is the host's name as a string data type
def check_host(host) :
    check = df[df['host_name']== host]
    total = check['name'].count()
    return f"{host} has {total} number of properties"
```

```python
#Testing the function to see if it works
check_host('Jennifer')
```
